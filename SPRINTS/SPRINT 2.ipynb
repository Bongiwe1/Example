## STOCK DATA ANALYSIS FOR ZAMA STORE
>OBJECTIVE:
main purpose of the analysis is maximize profit made for the business and its shareholders, by tracking whether profit is mostly generated from the stock quantity or the selling price.

```python
import pandas as pd
import numpy as np
import pymysql
import sqlalchemy
import matplotlib.pyplot as plt

#connection to database
engine = sqlalchemy.create_engine("mysql+pymysql://root:Bphavi04,@localhost/bongiwe")
df = pd.read_sql_table("stock_data1", engine)
df
```

```python
#create line graphs for stock difference
x = df["Stock Item"]
y1 = df["Stock price"]
y2 = df["Stock Selling Price"]

fig = plt.figure(figsize=(15, 5))
plt.xticks(rotation=45)
plt.plot(x, y1, 'gh-', label = "Stock price")
plt.plot(x, y2, 'bh-', label = "Stock Selling Price")

# Plot axes labels and show the plot
plt.title("Stock Price vs Selling Price", size = 15)
plt.xlabel('Items', size = 12)
plt.ylabel('Prices', size = 12)
plt.legend()

plt.show()
```

```python
# Creating a scatter plot
x1 = df["Stock price"]
y2 = df["Stock Selling Price"]

#define color and size of each data point

plt.scatter(x1, y2, c= "red", s= 10, marker = "*")


# Labeling the plot and showing it
plt.title("Stock price vs Selling price")
plt.xlabel("Stock Price")
plt.ylabel("Selling price")

plt.show()
```

```python
#Histogram for total number of stock
y = df["Quantity"]

# Creating a histogram with edge color to divide bins
plt.hist(y, edgecolor = "red", bins = [20, 25, 30, 35, 40, 45, 50, 55])

plt.title("Stock Quantity")
plt.ylabel('Frequency')

plt.show()
```

```python
df["Expected Profit"] = df["Stock Selling Price"] - df["Stock price"]
```

```python
df["Expected Profit"]
```

```python
fig = plt.figure(figsize=(15, 5))
plt.xticks(rotation=45)

x = df["Stock Item"]
y = df["Expected Profit"]

plt.bar(x, y)

plt.title("Expected Profit", size = 14)
plt.xlabel("Items", size = 14)
plt.ylabel("Price (R)", size = 14)

plt.show
```

```python
# Creating a scatter plot for Quantity
x1 = df["Quantity"]
x2 = df["Stock Selling Price"]
y2 = df["Expected Profit"]

plt.scatter(x1, y2, c= "purple", s= 20, marker = "*")

# Labeling the plot and showing it
plt.title("Quantity vs Profit")
plt.xlabel("Quantity")
plt.ylabel("Expected Profit")

plt.show()

#second scatter plot for Selling price
plt.scatter(x2, y2, c= "green", s= 20, marker = "*")

# Labeling the plot and showing it
plt.title("Selling price vs Profit")
plt.xlabel("Selling price")
plt.ylabel("Expected Profit")

plt.show()
```
